```python
import pandas as pd
import json
```

```python
with open('t20_json_files/t20_wc_match_results.json')as f:
    data=json.load(f)
#creating dataframe 
#in dataframe we can supply entire list of array i.e data[0]
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match
```

```python
#use scorecard as key i.e match id so it will be easy to connect with other table becase in scorecarsd each value is unique 
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
df_match
```

```python
match_ids_dict={}

for index, row in df_match.iterrows():
    key1=row['team1'] + 'Vs' + row['team2']
    key2=row['team2'] + 'Vs' + row['team1']
    match_ids_dict[key1]=row["match_id"]
    match_ids_dict[key2]=row["match_id"]
match_ids_dict
```
